```python
import xml.etree.ElementTree as ET
import pandas as pd
liste1 = []
for file in L:
    tree = ET.parse(file)
    root = tree.getroot()
    rows1 = []
    dossier_row = {}
    for child in root:
        if child.tag == 'Dossier':
            for sub_child in child:
                dossier_row[sub_child.tag] = sub_child.text
            rows1.append(dossier_row.copy())
            liste1.append(rows1)


df1 = pd.DataFrame(liste1)


liste11 = []
for index, row in df1.iterrows():
    for k,v in row.items():
        liste11.append(v)

u=[]
for i in liste11:
    dct = {k:v for k,v in i.items()}
    u.append(dct)
g = pd.DataFrame(u)



liste2 = []
for file in L :
    tree = ET.parse(file)
    root = tree.getroot()
    rows2 = []
    row2 = {}
    for child in root:
        if child.tag == 'Decision':
            for sub_child in child:
                row2[sub_child.tag] = ''.join([p.text.strip() for p in sub_child.findall('p')])
            rows2.append(row2)
            liste2.append(rows2)

pd.set_option("max_colwidth", None)
df2 = pd.DataFrame(liste2)


liste22 = []
for index, row in df2.iterrows():
    for k,v in row.items():
        liste22.append(v)


r=[]
for i in liste22:
    dct = {k:v for k,v in i.items()}
    r.append(dct)
m = pd.DataFrame(r)


merged_df= g.merge(m,right_index=True , left_index=True )
display(merged_df)
```
